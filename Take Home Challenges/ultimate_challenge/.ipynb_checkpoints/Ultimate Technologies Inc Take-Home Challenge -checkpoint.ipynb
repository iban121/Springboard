#### Aggregate these login counts based on 15minute time intervals, and visualize and describe the resulting time series of login counts in ways that best characterize the underlying patterns of the demand.

First let's set up a count for the logins. Then we can aggregate these. 

```python
data = data.set_index('login_time')
```

```python
data['count'] = 1
```

```python
data.head(2)
```

```python
data.tail(2)
```

```python
df_15_min = data.resample('15min').sum()
```

```python
df_15_min.head(2)
```

```python
_ = df_15_min.plot(kind='line')
_ = plt.gcf().set_size_inches(15,6)
_ = plt.show()
# plotly graph as it's easier to read off the values 
fig = px.line(df_15_min, x= df_15_min.index, y= 'count', width = 800)
fig.show()
```

```python
df_15_min.describe()
```

```python
df_day = data.resample('24H').sum()
_ = df_day.plot(kind='line')
_ = plt.gcf().set_size_inches(15,6)
_ = plt.show()
fig = px.line(df_day, x= df_day.index, y= 'count', width = 800)
fig.show()
```

```python
df_day['name_of_day'] = df_day.index.day_name()
df = df_day.groupby('name_of_day').sum()
```

```python
_ = df.plot(kind='line')
_ = plt.gcf().set_size_inches(15,6)
_ = plt.show()
```

```python
df_month = data.resample('M').sum()
_ = df_month.plot(kind='line')
_ = plt.gcf().set_size_inches(15,6)
_ = plt.show()
```
