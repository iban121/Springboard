# Aim: Identify which factor predicts future user adoption

#### Adopted User = user who logs in on three spearate days in a 7 day period

So before we do anything further, let's set up a column which identifies the users as adopted or not

```python
df_engagement.head(1)
```

```python
df_engagement = df_engagement.set_index('time_stamp')
```

```python
df_engagement
```

```python
from datetime import timedelta

def adopted(user):    
    "takes a users input and returns whether or not they have been active within any 7-day period"
    users = df_engagement.loc[df_engagement['user_id'] == user] #Look for object_id = x
    users = users.resample('D').mean().dropna() #resample for day
    adopted = 0
    for i in range(len(users)-2): #for loop so we can loop over the days until the second last day
        if users.index[i + 2] - users.index[i] <= timedelta(days=7): # difference between every 1st and 3rd day must be 7 days or less
            adopted = 1
            break
        else:
            adopted = 0
    return adopted
    
```

```python
df_users['adopted'] = df_users['object_id'].apply(adopted)
```

```python
df_users.head(2)
```

```python
df_engagement.head(2)
```

```python
df_users['adopted'].value_counts()
```

This tells us that 1656 out of the 8823 unique users are adopted users.

## Scaling and Classification

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, matthews_corrcoef, roc_auc_score, roc_curve, f1_score, recall_score
from sklearn.model_selection import GridSearchCV, cross_validate
```

```python
X = df_users.drop('adopted', axis = 1)
y = df_users['adopted']
X_train, X_test, y_train, y_test = train_test_split(X, np.ravel(y), test_size = 0.2, random_state = 10)
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Metrics Table 

```python
column_names = ['Classifier', 'Parameters' ,'F1', 'MCC', 'Accuracy', 'ROC_AUC', 'Recall',]
models = pd.DataFrame(columns= column_names)
def metrics_table(model_name, model):
    prediction = model.predict(X_test)
    summary = {'Classifier' : model_name,
               'F1': f1_score(y_test,prediction),
               'MCC': matthews_corrcoef(y_test, prediction),
               'Accuracy': accuracy_score(y_test, prediction),
               'ROC_AUC': roc_auc_score(y_test, prediction),
               'Recall' : recall_score(y_test, prediction),
               'Parameters': model.best_params_
        }
    return models.append(summary, ignore_index=True)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
parameters = {'n_estimators':np.arange(1,10), 'max_depth': np.arange(1,10)}
model_rfc_grid = RandomForestClassifier(random_state = 10)
model_rfc_grid_cv = GridSearchCV(model_rfc_grid, parameters, cv=10)
model_rfc_grid_cv.fit(X_train,y_train)
print('Best Score: ', model_rfc_grid_cv.best_score_)
print('Best parameters: ', model_rfc_grid_cv.best_params_)
```

```python
model_rfc_best = RandomForestClassifier(n_estimators = 9, max_depth=9, random_state = 10)
model_rfc_best.fit(X_train, y_train)
predictions_rfc_best = model_rfc_best.predict(X_test)
# fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
y_pred_prob = model_rfc_best.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
auc = roc_auc_score(y_test, y_pred_prob)
plt.figure(figsize=(10,6))
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.savefig('roc_best_random.png', bbox_inches = 'tight')
plt.show()
print('Area Under Curve: ',auc)
```

```python
sorted_importances = model_rfc_best.feature_importances_.argsort()
_ = plt.figure(figsize=(10,10))
_ = plt.barh(df_users.drop('adopted',axis = 1).columns[sorted_importances], model_rfc_best.feature_importances_[sorted_importances])
_ = plt.title('Feature Importance using GridSearch for Hyperparameter Tuning')
_ = plt.savefig('feature_imp_grid.png', bbox_inches='tight')
plt.show()
```

```python
rfc = metrics_table(model_name = 'Random Forest Classifier',
             model = model_rfc_grid_cv)
```
