# 2. Data Wrangling 

We want to do two things here: <br>
> 1. Remove the percentage signs <br>
> 2. Change data type to float for everything barring the index and beverage and beverage_prep columns. 

```python
percent_columns = ['Vitamin A (% DV) ', 'Vitamin C (% DV)', ' Calcium (% DV) ', 'Iron (% DV) ']
```

```python
df[percent_columns] = df[percent_columns].replace({'%':''}, regex = True)
```

```python
object_columns = [' Total Fat (g)']
```

```python
df['Vitamin A (% DV) '] = df['Vitamin A (% DV) '].astype(float)
df['Vitamin C (% DV)'] = df['Vitamin C (% DV)'].astype(float)
df[' Calcium (% DV) '] = df[' Calcium (% DV) '].astype(float)
df['Iron (% DV) '] = df['Iron (% DV) '].astype(float)
```

There are rows where the Caffeine entry is 'Varies' and in Total Fat (g) there are values separated by spaces. We need to investigate this further and decide how to handle these values. 

## 2.1 Data Imputation

```python
df[df['Beverage'].str.contains('Iced Tazo')]
```

We notice that beverages classified as Tazo®Tea drinks have zero values for all columns. As these will be hard to impute with any reliability, we'll drop these values. 

```python
df = df[df.Beverage != 'Tazo® Tea']
```

For shaken iced beverages, we do have data for the other columns. So for these, first let's change the 'Varied' to NaN and then we can impute these with the mean caffeine values. 

```python
df = df.replace('Varies', np.nan)
df = df.replace('varies', np.nan)
```

```python
df = df.fillna(df.mean())
```

```python
df['Caffeine (mg)'] = df['Caffeine (mg)'].astype(float)
```

For the Total Fat (g) column, there happens to be an entry with 3 2 which looks like should be 3.2, so we'll go ahead and replace this value. 

```python
df[df[' Total Fat (g)']== '3 2'] = df[df[' Total Fat (g)']== '3.2']
```

```python
df[' Total Fat (g)'] = df[' Total Fat (g)'].astype(float)
```

```python
df = df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
df[df['Beverage'].isnull()] 
```

We can't really classify the missing entry as any known type of drink nor by beverage prep, so it's best to drop this entry. 

```python
df = df.dropna(axis = 0)
```

Let's now have a look at the White Chocolate Mocha by itself. 

```python
white_choc = df[df['Beverage'].str.contains('White Chocolate')]
white_choc.head()
```

## Calories
Let's look at the different categories of drinks to come to an understanding of which types of drinks are healthier than others. 

```python
_ = plt.figure(figsize=(10,6))
_ = sns.hi(x='Calories', y='Total_Carbohydrates(mg)', data=df, hue=df.index)
_ = plt.legend(bbox_to_anchor=[1.45,1])
_ = plt.title('Total Carbohydrates vs Calories')
_ = plt.show()
```

```python
_ = plt.figure(figsize=(10,6))
_ = sns.scatterplot(x='Calories', y='Cholesterol(mg)', data=df, hue=df.index)
_ = plt.legend(bbox_to_anchor=[1.45,1])
_ = plt.title('Cholesterol vs Calories')
_ = plt.show()
```

```python
_ = plt.figure(figsize=(10,6))
_ = sns.scatterplot(x='Calories', y='Sugars(mg)', data=df, hue=df.index)
_ = plt.legend(bbox_to_anchor=[1.45,1])
_ = plt.title('Cholesterol vs Calories')
```

We can see that there are some patterns that are consistent throughout the three plots. <br>
> 1. Blended Coffee Frappuccino almost mark out one of the steeper lines: this can be inferred as they are the drinks with higher sugar and carbohydrates, and as a result higher number of calories and high in cholesterol.  
> 2. The light blended frappuccinos are a little more scattered, and although, they show the general linear trend these are a little less tightly packet. They appear to have the highest carbohydrates content compared the other beverages.
> 3. Classic Espresso drinks seem to have the lowest calories, sugar content, and cholesterol, but they have a steep trend when considering calories and carbohydrates. 
> 4. Shaken Iced Beverages seem to behave in the opposite manner to the espresso drinks.
